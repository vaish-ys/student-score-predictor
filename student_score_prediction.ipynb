```python
import pandas as pd

data = {
    "hours_studied": [1,2,3,4,5,6,7,8],
    "sleep_hours": [8,7,6,6,5,5,4,4],
    "previous_score": [50,55,60,65,70,75,80,85],
    "final_score": [52,57,63,67,72,78,82,88]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[["hours_studied","sleep_hours","previous_score"]]
y = df["final_score"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
import pandas as pd

new_data = pd.DataFrame([[5,6,70]], columns=["hours_studied","sleep_hours","previous_score"])
prediction = model.predict(new_data)

print("Predicted Score:", prediction)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["hours_studied"], df["final_score"])
plt.xlabel("Hours Studied")
plt.ylabel("Final Score")
plt.title("Study Hours vs Score")

plt.show()
```

```python
import pickle

pickle.dump(model, open("student_model.pkl", "wb"))
```
